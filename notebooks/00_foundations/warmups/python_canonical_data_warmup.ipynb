## Practice Python Containers

Use records, lists, dictionaries, and helper functions before moving into pandas-heavy analysis.

```python
clients = learner_tables["clients"]
transactions = learner_tables["transactions"]

active_client_ids = [
    row["client_id"]
    for row in clients[["client_id", "status"]].to_dict("records")
    if row["status"] == "active"
]

amounts_by_transaction_type = {}
for row in transactions[["transaction_type", "amount_chf"]].to_dict("records"):
    transaction_type = row["transaction_type"]
    amounts_by_transaction_type.setdefault(transaction_type, 0)
    amounts_by_transaction_type[transaction_type] += row["amount_chf"]

assert active_client_ids
assert amounts_by_transaction_type

{
    "active_client_count": len(active_client_ids),
    "transaction_amounts_by_type": amounts_by_transaction_type,
}
```

```python
def describe_small_table(table_name, frame):
    """Return a compact Python dictionary for one generated table."""
    return {
        "table_name": table_name,
        "row_count": len(frame),
        "column_count": len(frame.columns),
        "first_columns": list(frame.columns[:4]),
    }


small_table_summaries = [
    describe_small_table(table_name, frame)
    for table_name, frame in learner_tables.items()
    if len(frame) <= 12
]

assert any(summary["table_name"] == "alerts" for summary in small_table_summaries)
small_table_summaries
```
